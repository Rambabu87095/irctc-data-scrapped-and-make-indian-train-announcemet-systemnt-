```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd

url='https://erail.in/trains-between-stations/patna-jn-PNBE/new-delhi-NDLS'
res=requests.get(url)
content=BeautifulSoup(res.content,'html.parser')
content1=content.prettify()
dd=content.find_all('div',class_='OneTrain')
```

```python
dd
```

```python

a=[]
for i in dd:
    a.append(i.text)
train_no_name=[]
for i in a:
    train_no_name.append(i.split('\n')[2])
train_no=[]
for i in train_no_name:
    train_no.append(i.split(' ',1))
Train_no=[]
for i in train_no:
    Train_no.append(i[0])
Train_no
```

```python
Train_name=[]
for i in train_no:
    Train_name.append(i[1])
Train_name
```

```python
j=[]
for i in dd:
    j.append(str(i))
#j
b=[]
for g in j:
    #print(g)
    b.append(g.split('_'))
#b
from_=[]
for i in b:
    from_.append(i[2])
From=[]
for i in from_:
    if i=='PNBE':
        From.append(i.replace('PNBE', "Patna"))
    elif i=='HJP':
        From.append(i.replace('HJP', "Hajipur"))
    elif i=='DNR':
        From.append(i.replace('DNR', "Danapur"))
    else:
        From.append(i.replace('PPTA', "Patliputra"))
From
```

```python
from datetime import date
x=date.today()
c=(x.strftime("%d"+"-"+"%b"))
c
```

```python
dep=content.find_all('div',{'class':'ArrDep bold Sorted'})

dep_ti=[]
for i in dep:
    dep_ti.append(i.text)
#dep_ti
Dep_date=[]
Dep_time=[]
for j in dep_ti:
    if len(j)==5:
        Dep_time.append(j)
        Dep_date.append(c)
    elif len(j)>5:
        Dep_time.append(j[:5])
        Dep_date.append(j[5:])
```

```python
destination=[]
for i in a:
    destination.append(i.split('\n')[3])
```

```python
to=''
travel_hr=''
for i in destination:
    for j in i:
        if j.isnumeric():
            travel_hr=travel_hr+j
        else:
            to=to+j         
to=to.split('.')
To=[]
for i in to:
    To.append(i.replace('hr',''))

To=To[:-1]  
To
```

```python
Arr_time=[]
Arr_date=[]
for i in b:
    Arr_time.append(i[7])
    
Arr_time
```

```python
for i in b:
    Arr_date.append(i[6][:6])
Arr_date
```

```python
Travel_hr=''
for j,i in enumerate(travel_hr):
    Travel_hr=Travel_hr+i
    if j==0:
        continue
    elif j==1:
        Travel_hr=Travel_hr+':'
    elif j==2:
        continue
    elif j==3:
        Travel_hr=Travel_hr+','  
    elif j%4==1:
        Travel_hr=Travel_hr+':'
    elif j%2==1:
        Travel_hr=Travel_hr+','
Travel_hr=Travel_hr.split(',')[:-1]
Travel_hr
```

```python
data=pd.DataFrame({'Train_no':Train_no,'Train_name':Train_name,'From':From,'Dep_date':Dep_date,'Dep_time':Dep_time,'To':To,'Arr_date':Arr_date,'Arr_time':Arr_time,'Travel_hr':Travel_hr})
data
```

```python
b=[]
c=''
for j,i in enumerate(data['Train_no']):
    j
    for i in data['Train_no'][j]:
        c=c+i+' '
    b.append(c)
    c=''
    
data['Train_no']=b
```

```python
a=[]
for i in data['Train_name']:
    i=i+'-'
    a.append(i.split())
b=[]
for i in a:
    for j in i:
        if len(j)>4:
            b.append(j.capitalize())
        else:
            b.append(j)
d=[]
for i in b:
    d.append(i.replace('EX-','Express-'))
e=[]
for i in d:
    e.append(i.replace('EXP-','Express-'))
f=''

for i in e:
    f=f+i+' '
h=f.split('-')
data['Train_name']=h[:-1]
```
